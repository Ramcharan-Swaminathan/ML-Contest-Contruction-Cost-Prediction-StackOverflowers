# Construction Dataset Analysis

This notebook provides a basic exploratory data analysis (EDA) of the construction dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Set plot style
plt.style.use('ggplot')
sns.set_theme(style="whitegrid")
```

## Loading the Dataset

```python
dataset_path = 'dataset/train_tabular.csv'
if os.path.exists(dataset_path):
    df = pd.read_csv(dataset_path)
    print("Dataset loaded successfully.")
else:
    print(f"Error: Dataset not found at {dataset_path}")
```

## Basic Exploratory Data Analysis

```python
# Display the first few rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Statistical summary
df.describe()
```

## Visualizing Missing Values

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

## Visualizing the Target Variable Distribution

Understanding the distribution of `construction_cost_per_m2_usd` is crucial as it is our target variable.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['construction_cost_per_m2_usd'], kde=True, bins=30, color='royalblue')
plt.title('Distribution of Construction Cost per m2 (USD)')
plt.xlabel('Cost (USD)')
plt.ylabel('Frequency')
plt.show()
```

## Univariate Analysis of Categorical Features

Let's look at how the cost varies across different categories like `country`, `seismic_hazard_zone`, etc.

```python
categorical_cols = ['country', 'developed_country', 'seismic_hazard_zone', 'flood_risk_class']

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
for i, col in enumerate(categorical_cols):
    sns.boxplot(x=col, y='construction_cost_per_m2_usd', data=df, ax=axes[i//2, i%2], palette='Set2')
    axes[i//2, i%2].set_title(f'Cost vs {col}')
    axes[i//2, i%2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Correlation Matrix

Checking correlations between numerical features.

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Trends over Time

Visualizing how the average cost changes over years/quarters.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='quarter_label', y='construction_cost_per_m2_usd', data=df.sort_values('quarter_label'), errorbar=None, marker='o')
plt.title('Average Construction Cost Trend over Quarters')
plt.xticks(rotation=45)
plt.ylabel('Avg Cost (USD)')
plt.show()
```
